```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(uri)
dados.head()
```

```python
mapa = {
    "home": "principal",
    "how_it_works": "como_funciona",
    "contact": "contato",
    "bought": "comprou"
}

dados = dados.rename(columns = mapa)
```

```python
x = dados[["principal","como_funciona","contato"]]
y = dados['comprou']
x.head()
```

# Dados de treino/teste
A primeira coisa se fazer é separar os dados do nosso algoritmo, dos dados de treino e os dados de teste. Não queremos que ele trabalhe com todos os dados. Se dermos todos os dados ele irá acertar 100% das vezes, porém nada fora do conjunto. Nós queremos dar uma  amostra de dados e treina-lo a prever.

```python
# Tamanho de 99  linhas e 4  colunas
dados.shape
```

```python
# Separando aproximadamente 25% para testar
# Separando a maior parte, 75% para treinar

treino_x = x[:75]
treino_y = y[:75]
teste_x = x[75:]
teste_y = y[75:]

print("Treinaremos com  %d elementos e testaremos com %d elementos" %(len(treino_x), len(teste_x)) )
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

# Usando a biblioteca para separar treino e teste

```python
from sklearn.model_selection import  train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.25)
print("Treinaremos com  %d elementos e testaremos com %d elementos" %(len(treino_x), len(teste_x)) )

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        random_state = SEED, test_size = 0.25,
                                                        stratify = y )
print("Treinaremos com  %d elementos e testaremos com %d elementos" %(len(treino_x), len(teste_x)) )

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```
